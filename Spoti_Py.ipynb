```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import snowflake.connector
import schedule
import time

# Spotify API credentials
spotify_client_id = '921952ab6b294674b236bab262e26186'
spotify_client_secret = 'ee90cfc8fd7b470d976732e61ed90675'

# Snowflake credentials
snowflake_username = ''
snowflake_password = '@90665'
snowflake_account = ''
snowflake_database = 'DEMO'
snowflake_schema = 'PUBLIC'
snowflake_table = ''

# Spotify playlist ID
playlist_id = ''  # Replace with your desired playlist ID

def extract_and_load_spotify_data():
    # Establish a connection to Spotify
    spotify_auth = SpotifyClientCredentials(
        client_id=spotify_client_id,
        client_secret=spotify_client_secret
    )
    sp = spotipy.Spotify(client_credentials_manager=spotify_auth)

    # Extract data from Spotify
    results = sp.playlist_tracks(playlist_id)

    # Create a list to store the extracted data
    spotify_data = []

    for track in results['items']:
        track_name = track['track']['name']
        artist_name = track['track']['artists'][0]['name']
        spotify_data.append((track_name, artist_name))

    # Establish a connection to Snowflake
    conn_snowflake = snowflake.connector.connect(
        user=snowflake_username,
        password=snowflake_password,
        account=snowflake_account,
        #warehouse='YOUR_WAREHOUSE',  # Replace with your Snowflake warehouse
        database=snowflake_database,
        schema=snowflake_schema
    )

    # Create a Snowflake cursor
    cur_snowflake = conn_snowflake.cursor()

    # Truncate the table (optional, to remove existing data)
    truncate_query = f"TRUNCATE TABLE {snowflake_table}"
    cur_snowflake.execute(truncate_query)

    # Insert data into Snowflake table
    insert_query = f"INSERT INTO {snowflake_table} (TRACK_NAME, ARTIST_NAME) VALUES (%s, %s)"
    cur_snowflake.executemany(insert_query, spotify_data)

    # Commit the transaction
    conn_snowflake.commit()

    # Close Snowflake connection
    conn_snowflake.close()

# Schedule the extraction function to run every hour
schedule.every(10).minutes.do(extract_and_load_spotify_data)

# Run the scheduled tasks indefinitely
while True:
    schedule.run_pending()
    time.sleep(1)
```
