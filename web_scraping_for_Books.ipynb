```python
#hence there are 20 books on one page.
page=1 #to count number of pages of the website
titles=[] # list for titles of all the books
valid_page=True

while valid_page: #loop for all the web pages 
    url1=base_url.format(page)
    res2=requests.get(url1)
          
    if '404 Not Found' in res2.text:
        break;
    
    soup1=bs4.BeautifulSoup(res2.text,'lxml')
    
    for i in range(len(soup1.select('h3'))): #loop for all the number of books oon one page
        titles.append(soup1.select('h3 a')[i]['title'])
        
        
        
    page=page+1
```

```python
#now we need all of this for all the 1000 books.

#for that we again need a loop
#now while surfing through som of the pages we can see the pattern in url. Every url has the name of the book with its 
#position in the list.
#so we can collect the links to different books' individual pages. as we saw above that each page has 20 books. Hence we 
#will collect the link o 20 books from all the web pages.

#in total we will be having 1000 links.
base_url.format(50)
```

```python
res_books_on_single_page=requests.get(base_url.format(50))
soup_books_on_single_page=bs4.BeautifulSoup(res_books_on_single_page.text,'lxml')
```

```python
books_on_a_single_page=[]
for i in range(len(soup_books_on_single_page.select('h3'))):
        books_on_a_single_page.append(soup_books_on_single_page.select('h3 a')[i]['href'])
```

```python
#Similary we can have links for all the books by looping the above for loop for all the number of pages of the website
link_for_individual_book=[]
for i in range(51):

        
    res_books_on_single_page=requests.get(base_url.format(i))
    soup_books_on_single_page=bs4.BeautifulSoup(res_books_on_single_page.text,'lxml') 
    books_on_a_single_page=[]
    for i in range(len(soup_books_on_single_page.select('h3'))):
        link_for_individual_book.append(soup_books_on_single_page.select('h3 a')[i]['href'])
```

```python
#when we will look closely we will see that this is the common part of the link
base_url1='http://books.toscrape.com/catalogue/'
```

```python
category=[]
price=[]
price_after_tax=[]
tax_amnt=[]
availability=[]
num_reviews=[]
description=[]
img_link=[]
star_rating=[]
for i in range(1000):
    
       
    product_url=base_url1+link_for_individual_book[i]
    
    res_product=requests.get(product_url)
    soup_product=bs4.BeautifulSoup(res_product.text,'lxml')
    #category
    category.append(soup_product.select('.breadcrumb li a')[2].text)
    
    #price
    price.append(soup_product.select('.table.table-striped td')[2].text)
    #price after tax
    price_after_tax.append(soup_product.select('.table.table-striped td')[3].text)
    #tax amnt
    tax_amnt.append(soup_product.select('.table.table-striped td')[4].text)
    #availability
    availability.append(soup_product.select('.table.table-striped td')[5].text)
    #no. of reviews
    num_reviews.append(soup_product.select('.table.table-striped td')[6].text)
    #Product description
    description.append(soup_product.select('.product_page p')[3].text)
    #for the image link of books
    x=soup_product.select('.item.active img')[0]
    image_link='http://books.toscrape.com/'+x['src'][6:]
    img_link.append(image_link)
    #rating
    star_rating.append(soup_product.select('.col-sm-6.product_main p ')[2]['class'][1])
```

```python
import numpy as np
import pandas as pd
df=pd.DataFrame({'Title':titles,'Category':category,'Price':price,'Price_After_Tax':price_after_tax,'Tax_amount':tax_amnt,'Availability':availability,'Number_of_reviews':num_reviews,'Book_Description':description,'Image_Link':img_link,'Stars':star_rating})
df.tail()
```

```python
df.Image_Link.head()
```

```python
#some columns need cleaning ==>Data Cleaning
df.Price.head()
```

```python
def handle_different_amounts(value):
    str(value)
    value=value[2:]
    return float(value)
df['Price']=df.Price.apply(handle_different_amounts)
df.head()
```

```python
df['Price_After_Tax']=df.Price_After_Tax.apply(handle_different_amounts)
df['Tax_amount']=df.Tax_amount.apply(handle_different_amounts)
df.head()
```

```python
def handle_rating(value):
    str(value)
    if value=='One':
        value=1
    elif value=='Two':
        value=2
    elif value=='Three':
        value=3
    elif value=='Four':
        value=4
    elif value=='Five':
        value=5
    else:
        value=0
    return value
df['Stars']= df.Stars.apply(handle_rating)
df.head()
```

```python
def handle_availability(value):
    str(value)
    
    value= value.split(' ')[2][1:]
    return int(value)
df['Availability']=df.Availability.apply(handle_availability)
df.head()
```

```python
df.info()
```

```python
df['Number_of_reviews']=pd.to_numeric(df['Number_of_reviews'])
df['Price']=pd.to_numeric(df['Price'])
df.head()
```
